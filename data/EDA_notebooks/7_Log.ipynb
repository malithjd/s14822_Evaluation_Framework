<p style="background-color:#A0AE9A;color:#010521;font-size:200%;text-align:center;">Un Necessary Columns</p>

```python
df.drop(['Row ID','Order ID','Customer ID','Customer Name'],axis=1,inplace=True)
```

<p style="color:#544F78;
          font-size:22px;"> -> Unnecessary columns droped </p>

<p style="background-color:#A0AE9A;color:#010521;font-size:200%;text-align:center;">Remove Duplicates</p>

```python
df[df.duplicated()]
```

-------------

```python
df.drop_duplicates(inplace=True)
```

<p style="color:#544F78;
          font-size:22px;"> -> Duplicated rows droped </p>

<p style="background-color:#A0AE9A;color:#010521;font-size:200%;text-align:center;">Null Data</p>

```python
df.isnull().sum()
```

--------------------

```python
# show all the null row in the postalcode column 
df[df['Postal Code'].isnull()]
```

----------------------------

```python
#all missed postal codes is in america which equal 05401
df['Postal Code'].fillna(5401,inplace=True)
```

-----------------------------

```python
df.isnull().sum().sum()
```

<p style="color:#544F78;
          font-size:22px;"> -> Null data handled </p>

<p style="background-color:#A0AE9A;color:#010521;font-size:200%;text-align:center;">Columns DataTypes Validation</p>

```python
#validate columns types 
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

<p style="color:#544F78;
          font-size:22px;"> -> Columns types is valid </p>

Data visualization & getting conclusions

```python
top_cities= df.groupby("City").sum().sort_values("Sales",ascending=False).head(15)
top_cities= top_cities[['Sales']].round(2)
top_cities.reset_index(inplace=True)
top_cities

fig=px.bar(top_cities,x='City',y='Sales',title="Top 15 Cities",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> New york city have the most sales </p>

----------------

```python
top_states= df.groupby("State").sum().sort_values("Sales",ascending=False).head(15)
top_states= top_states[['Sales']].round(2)
top_states.reset_index(inplace=True)
top_states
fig=px.bar(top_states,x='State',y='Sales',title="Top 15 Cities",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> California have the most sales </p>

-----------------------

```python
fig = px.pie(df, values=np.ones(9799), names='Ship Mode', title='Ship Mode',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Standard Class was the most common ship mode </p>

------------------

```python
fig = px.pie(df, values=np.ones(9799), names='Segment', title='Segment',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Most orders was from consumers </p>

-------------------

```python
fig = px.pie(df, values=np.ones(9799), names='Category', title='Category',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Most orders was of Office Suppliers </p>

-----------------------

```python
Categoris= df.groupby("Category").sum().sort_values("Sales",ascending=False).head(15)
Categoris= Categoris[['Sales']].round(2)
Categoris.reset_index(inplace=True)

fig = px.pie(Categoris, values='Sales', names='Category', title='Categoris sales',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Technological products have the most sales </p>

--------------------

```python
fig = px.pie(df, values=np.ones(9799), names='Sub-Category', title='Sub Category',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Most orders was of Binders </p>

-------------------

```python
regions= df.groupby("Region").sum().sort_values("Sales",ascending=False)
regions= regions[['Sales']].round(2)
regions.reset_index(inplace=True)

fig=px.pie(regions,names='Region',values='Sales',title="Regions Sales",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Most sales was from the West Region </p>

----------------------

```python
Segments= df.groupby("Segment").sum().sort_values("Sales",ascending=False)
Segments= Segments[['Sales']].round(2)
Segments.reset_index(inplace=True)

fig=px.pie(Segments,names='Segment',values='Sales',title="Segments Sales",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> Most sales was from the consumers </p>

-------------------

```python
plt.figure(figsize=(18,10))
top_cities= df.groupby(["City","Category"]).sum().sort_values("Sales",ascending=False).head(40)
top_cities= top_cities[['Sales']].round(2)
top_cities.reset_index(inplace=True)
sns.barplot(x='City',y='Sales',hue='Category',data=top_cities,palette='dark:salmon_r')
plt.xticks(rotation='vertical')
plt.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> As we see in the figure the most common ctegory in every city which help in markting in the common product for each city </p>

-------------------

```python
plt.figure(figsize=(18,10))
top_cities= df.groupby(["City","Segment"]).sum().sort_values("Sales",ascending=False).head(40)
top_cities= top_cities[['Sales']].round(2)
top_cities.reset_index(inplace=True)
sns.barplot(x='City',y='Sales',hue='Segment',data=top_cities,palette='dark:salmon_r')
plt.xticks(rotation='vertical')
plt.show()
```

<p style="color:#544F78;
          font-size:22px;"> -> In the figure we can see the most common segment type in every city </p>
